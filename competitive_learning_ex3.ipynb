```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# numpy arr -> string
def np_str(arr):
  if len(arr.shape) == 1:
    return ', '.join([str(x) for x in arr])
  else:
    # column first order
    return ';\t'.join(['col%d: [%s]' % (i, np_str(x)) for i, x in enumerate(arr.T)])
```

```python
# inputs (column vectors)
x1 = np.random.randn(2, 100) / 10 + np.array([[5], [0]])
x2 = np.random.randn(2, 100) / 10 + np.array([[5], [5]])
x3 = np.random.randn(2, 100) / 10 + np.array([[-5], [-5]])
x4 = np.random.randn(2, 100) / 10 + np.array([[0], [5]])
x = np.hstack((x1, x2, x3, x4))
x /= np.linalg.norm(x, axis=0)

# inputs order (column ids)
order = list(range(len(x))) * 10

# initial weight vector (column vectors = neurons)
#w = np.array([[3, 0], [3, 3], [-3, -3], [0, 3.]]).T
#w /= np.linalg.norm(w, axis=0)
w = np.random.randn(2, 4) * 0.5
w /= np.linalg.norm(w, axis=0)

# learning rate
eta = .1
```

```python
def classify_all():
    return np.array([get_winner(t) for t in x.T])
```

```python
def get_winner(inp):
    # neuron activations
    activation = w.T @ inp

    # winner
    winner = np.argmax(activation)
    
    return winner
```

```python
def plot_clustering():
    xyc_arr = []
    color_map = ['blue', 'red', 'magenta', 'y']
    for x_ in np.linspace(-2, 2, 20):
        for y_ in np.linspace(-2, 2, 20):
            xyc_arr.append([x_, y_, color_map[get_winner([x_, y_])]])
    #plt.scatter(**dict(zip(['x', 'y', 'c'], zip(*xyc_arr))))
    plt.scatter(x[0], x[1], s=100, c='green')
    plt.scatter(w[0], w[1], s=100, c='black')
    plt.show()
```

```python
# going through all inputs
res = classify_all()
plot_clustering()
np.random.shuffle(order)
for j, i in enumerate(order):
  # current input
  inp = x[:, i]

  # neuron activations
  activation = w.T @ inp + np.random.randn(w.shape[1]) / (1 + j)

  # winner
  winner = np.argmax(activation)

  # winner update rule
  w[:, winner] += eta * (inp - w[:, winner])

  # renormalizing
  w /= np.linalg.norm(w, axis=0)
    
  # printing result
  print('Input %d: %s,\tactivation %s,\twinner %d,\tweights %s' % (i, np_str(inp), np_str(activation), winner, np_str(w)))
  new_res = classify_all()
  print(len(np.where(new_res != res)[0]))
  plot_clustering()
  res = new_res
```
